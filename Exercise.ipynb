#### b) [10%] Basierend auf den test-Fällen, schreiben Sie eine Methode welche den Task erfolgreich (und effizient) löst.

```python
## Task 2b – Implementation

The method find_single() was implemented to efficiently find the element that appears exactly once.

Implementation idea:

The solution uses the XOR operator.

Properties:

* a ^ a = 0
* a ^ 0 = a

Since every element except one appears exactly twice, all duplicate elements cancel each other out. The remaining value is therefore the unique element.

Algorithm:

"""
def find_single(numbers):
    result = 0

    for number in numbers:
        result ^= number

    return result
"""

Complexity:

* Time complexity: O(n)
* Space complexity: O(1)

This approach is more efficient than counting occurrences with nested loops and does not require additional memory structures.
```
